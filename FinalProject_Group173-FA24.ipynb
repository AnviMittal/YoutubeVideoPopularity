## Trending YouTube Videos Dataset

### Cleaning & Processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

filenames = ['./datasets/countries/USvideos.csv', './datasets/countries/RUvideos.csv', './datasets/countries/GBvideos.csv', './datasets/countries/MXvideos.csv', './datasets/countries/KRvideos.csv', './datasets/countries/JPvideos.csv', './datasets/countries/INvideos.csv', './datasets/countries/FRvideos.csv', './datasets/countries/DEvideos.csv', './datasets/countries/CAvideos.csv']
labels = ['US', 'RU', 'GB', 'MX', 'KR', 'JP', 'IN', 'FR', 'DE', 'CA'] # labels to separate dataset by country

# for loop for reading CSV files and putting them into a list of dataframes
dfs = []
for file, label in zip(filenames, labels):
    df = pd.read_csv(file, encoding='latin-1') # need 'latin-1' in order to read the russian csv
    df['source'] = label  # this is for accessing each dataset individually after merging them
    dfs.append(df)
```

```python
# merging the dataframes
df_trending_videos = pd.concat(dfs, ignore_index=True) 
```

```python
# dropping unnecessary columns
columns_to_drop = ['video_id', 'trending_date', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed','channel_title','thumbnail_link']
df_trending_videos = df_trending_videos.drop(columns=columns_to_drop)
df_trending_videos = df_trending_videos[df_trending_videos['comment_count']>11]
df_trending_videos = df_trending_videos.dropna()

#Creating column for more dislikes than likes
df_trending_videos['dislikes_over_likes'] = df_trending_videos['dislikes'] > df_trending_videos['likes']
```

```python
# creating subset dataframes so each country can be easily accessed
us_data = df_trending_videos[df_trending_videos['source'] == 'US']
ru_data = df_trending_videos[df_trending_videos['source'] == 'RU']
gb_data = df_trending_videos[df_trending_videos['source'] == 'GB']
mx_data = df_trending_videos[df_trending_videos['source'] == 'MX']
kr_data = df_trending_videos[df_trending_videos['source'] == 'KR']
jp_data = df_trending_videos[df_trending_videos['source'] == 'JP']
in_data = df_trending_videos[df_trending_videos['source'] == 'IN']
fr_data = df_trending_videos[df_trending_videos['source'] == 'FR']
de_data = df_trending_videos[df_trending_videos['source'] == 'DE']
ca_data = df_trending_videos[df_trending_videos['source'] == 'CA']
```

```python
from datetime import datetime
import pytz

utc_tz = pytz.utc
pst_tz = pytz.timezone("US/Pacific")

#Function to convert time to PST
def convert(time):
    utc_time = datetime.strptime(time, "%Y-%m-%dT%H:%M:%S.%fZ").replace(tzinfo=utc_tz)
    pst_time = utc_time.astimezone(pst_tz)
    # Format to military (24-hour) time and exclude the year
    return pst_time.strftime("%H:%M PST")

#Creating a column for publish time
df_trending_videos['publish_time'] = df_trending_videos['publish_time'].apply(convert)

#Creating a column for engagement rate
df_trending_videos.loc[:, 'engagement_rate'] = (
    (df_trending_videos['likes'] + df_trending_videos['comment_count']) / df_trending_videos['views']
)

#Creating a column for title length
df_trending_videos['title_length']= df_trending_videos['title'].str.replace(' ','').str.len()
```

```python
# Created 2 datasets for >100K and <100K
df_greater_100k = df_trending_videos[df_trending_videos['views']>=100000]
df_less_100k = df_trending_videos[df_trending_videos['views']<100000]
```

### Relationship between number of views and publish time of a video

This bar plot has been generated to analyze the relationship between the time of the day a video is published and the number of views it receives. In order to achieve this, we created 4 different publish periods : first,second,third and fourth and populated a new column with these values for each video based on the publish time. We plotted the total number of views received by videos in each publish period. The graph clearly shows that videos posted in the second publish period(6AM-11:59AM PST) receive the highest views while videos posted in the fourth publish period(6PM-11:59PM PST) received the least number of total views. This analysis shows that publishing a video during the second time period can maximize views.

```python
#Views based on time of day

def publish_period(time):
    h = int(time.split(':')[0])
    if 0 <= h < 6:
        return "first"
    elif 6 <= h < 12:
        return "second"
    elif 12<= h < 18:
        return "third"
    else:
        return "fourth"
    
#Creating time period column in both datasets

df_trending_videos['time_period'] = df_trending_videos['publish_time'].apply(publish_period)

#dotring time period order so it appears chronoligally
time_period_order = ['first', 'second', 'third', 'fourth']
df_trending_videos['time_period'] = pd.Categorical(df_trending_videos['time_period'], categories=time_period_order, ordered=True)
#df_less_100k['time_period'] = pd.Categorical(df_less_100k['time_period'], categories=time_period_order, ordered=True)

#finding total videos published at a vertain time
total_vids= df_trending_videos.groupby('time_period', observed=False).size()

color_map = {'first': 'skyblue', 'second': 'red', 'third': 'green', 'fourth': 'purple'}

total_vids.plot(kind='bar', color = [color_map[period] for period in total_vids.index], legend= False)
plt.xlabel('Publish Time Period')
plt.ylabel('Total views')
plt.title('Total views based on when video was published')
legend_labels = [
    " 12:00 AM - 5:59 AM",
    " 6:00 AM - 11:59 AM",
    " 12:00 PM - 5:59 PM",
    " 6:00 PM - 11:59 PM"
]
for i, color in color_map.items():
    plt.plot([],[], color = color, label = f'{i}:{legend_labels[time_period_order.index(i)]}')

plt.legend(title="Time Periods", loc='upper right');


```

# Discussion and Conclusion

By employing methods of data wrangling and exploratory data analysis on a dataset of Youtube videos, this research project aims to assess the influence of variables like video category, title length, publish time, use of tags, and engagement metrics on the number of views a YouTube video receives. This dataset was cleaned by eliminating redundant columns and null/undefined values and by creating new relevant columns to standardize data for efficient analysis.This dataset was split based on the threshold value of 100k views.  By comparing these variables for datasets containing videos with less than a 100k views and more than a 100k views, we attempt to explain the factors that play a role in achieving more than a 100k views.

Generating a graph to compare the engagement metrics of videos with less than 100k and more than 100k views reveals that those receiving more than 100k views consistently demonstrate higher engagement than the alternative. The presence of multiple outliers indicates that a high frequency of videos receiving large viewership display high engagement metrics. There is a very limited number of low reach videos that receive high engagement. Thus, there is a positive correlation between the number of views and engagement metrics. This is consistent with the findings of Broxton et al(2013), which suggest that user interaction has a positive influence on viewership. This highlights the importance of high engagement through likes and comments in reaching a larger audience. These findings are consistent with the hypothesis that states that engagement metrics have an impact on the number of views a video receives.

We wanted to analyze the relationship between the number of dislikes and high viewership count. We assessed this relationship as several YouTube videos with potentially offensive or controversial content tend to have more dislikes than likes, yet still achieve high viewership. We wanted to explore whether there is a connection between increased dislikes and view count. This was achieved by creating a new column that indicated through boolean values (true/false) whether a video's dislikes were more than its likes. After plotting the distribution of this column across both the datasets, one with more than 100k views and the other with less than 100k views, we observed that a higher number of videos received more likes than dislikes across both plots by a significant amount. This shows that receiving more dislikes than likes does not influence viewership in any case.  

We generated a histogram in order to analyze the influence of title length on the number of views that a video receives. The results show that a significant number of videos that have received more than a 100k views have a title length of less than 100 characters. This demonstrates that videos that reach a larger audience tend to have shorter titles. There are fewer videos with less than 100k views that maintain the same short title length as videos receiving more than a 100k views. These results highlight the importance of maintaining a title length of less than a 100 characters will boost viewership, as these trends are displayed by videos with more than a 100k views. This is consistent with the findings in the paper by Gong et al(2015), that suggested title optimization is a key factor in increasing the number of views. Thus, creators should create titles that have a shorter title length in order to reach a larger audience. These findings are inconsistent with our hypothesis, which suggests that video title length does not play a role in viewership.

In order to study the genre of YouTube videos that receive the highest number of views, we created a histogram for both datasets. The plot suggested that videos related to ‘Entertainment’ received the highest number of views in both datasets, indicating that this is the most popular genre for video creation. Additionally, ‘Music’ and ‘Comedy’ were also among the top categories associated with significant view counts. This suggests that one’s chances of increased viewership are highest when publishing entertainment or music-related content. These conclusions align with the findings of Hoiles et al. (2017), who demonstrated that ‘Entertainment’ is the most popular genre with the widest reach. However, while Hoiles et al. identified ‘Tutorials’ as the second most popular category, our analysis found ‘Music’ and 'Comedy' to be more prevalent. This also contradicts our initial hypothesis, which stated that content category had no relationship with view counts. These findings show the importance of selecting popular categories, like ‘Entertainment’ and ‘Music,’ for achieving higher viewership.

In order to study the relationship between the number of tags and video view counts, we created a histogram that displays the average number of views based on the number of tags a video contains. We decided to analyze this relationship because tags play a crucial role in enhancing a video's discoverability on YouTube. Proper tagging helps categorize the content better, making it easier for a user to find the video. Understanding the optimal number of tags can allow content creators to maximize their reach and engagement. The results indicate that videos with a moderate number of tags, specifically in the range of 5 to 35 tags, achieve the highest viewership. This trend is consistent for both our datasets. Conversely, the data reveals that the viewership starts to decrease when the tag count exceeds 60 tags. These findings suggest that using 5 to 35 tags increases the likelihood of higher viewership, while excessive tagging can reduce a video's reach.

We generated a bar plot to analyze the relationship between the publish time of a video and the number of views. The plots suggested that videos that have been posted between 6:00AM and 11:50AM gain the most popularity and receive the largest number of views. This could be attributed to factors like viewers being more active during that period of time. Understanding these intricacies can help achieve a larger viewership on the platform. These findings are consistent with the research by Sabate et al.(2014) that demonstrates that post timing has an influence on engagement and reach of a post. This highlights that videos posted between 6AM-11:59AM receive the maximum number of views.

In conclusion, various factors are at play in determining the popularity of a YouTube video. We found that factors such as engagement metrics, title length, and video genre have the most influence on a video achieving maximum views. As such, new creators beginning their YouTube journey should familiarize themselves with these metrics and the best way to use them. This will allow them to utilize YouTube recommendation algorithms for their own benefit to get the maximum possible reach for their videos.
